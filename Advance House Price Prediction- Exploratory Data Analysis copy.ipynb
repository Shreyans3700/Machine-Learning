## Missing Values

```python
#here we will check NaN values in each columns 

#step-1 - here we will make list of all the features with null values
features_with_na = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtype=='O']

#step-2 here we will calculate missing percentage of each values
for feature in features_with_na:
    print(feature,np.round(data[feature].isnull().mean(),4),'% missing values')
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    dataset=data.copy()
    dataset[features_nan]=dataset[features_nan].fillna('Missing')
    return data

data=replace_cat_feature(data,features_with_na)

data[features_with_na].isnull().sum()
```

### Since they are many missing values, we need to find the relationship between missing values and Sales Price

Let's plot some diagram for this relationship

```python
for feature in features_with_na:
    #copy the data into another variable
    dataset = data.copy()
    
    #let's make a variable that states 1 if feature value is missing and 0 if it's not
    dataset[feature] = np.where(dataset[feature].isnull(),1,0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    dataset.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

## Handling missing values

```python
#let's compare all NaN values
#first we will deal with categorical missing values 
categorical_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype=='O']

for feature in categorical_nan:
    print("{},{}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#replace new values with a new label
def replace_Categorical_Feature(dataset,features_nan):
    data = dataset.copy()
    data[features_nan] = data[features_nan].fillna("Missing")
    return data

dataset = replace_Categorical_Feature(dataset,categorical_nan)

dataset[categorical_nan].isnull().sum()
```

```python
dataset.head()
```

```python
#now we will deal with the numerical NaN Values
numerical_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtype!='O']

for feature in numerical_nan:
    print("{},{}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#replacing the numerical missing values 

for feature in numerical_nan:
    
    #finding the median value for the feature
    median = dataset[feature].median()
    
    #create a NEW FEATURE TO CAPTURE THE nan values
    dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0)
    
    #fitting the median value to nan
    dataset[feature].fillna(median,inplace=True)
    
dataset[numerical_nan].isnull().sum()
```

```python
dataset.head(50)
```

```python
#temporal variables(Date Variables)
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
dataset.head(10)
```

```python
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
# Since Some numerical features are skewed we will perform numerical transformation
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
    
dataset.head()
```

# Handling rare categorical features
- We will remove categorical variables that are present less than 1% of the observations

```python
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_feature
```

```python
for feature in categorical_feature:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset_feature = np.where(dataset[feature].isin(temp_df),feature,"Rare_Var")
```

```python
dataset.head(100)
```

```python
#converting categorical features to discrete labels
for feature in categorical_feature:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

# Feature Scaling

Feature Scaling is a technique to standardize the independent features present in the data in a fixed range. It is performed during the data pre-processing to handle highly varying magnitudes or values or units.

```python
feature_scale = [feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(dataset[feature_scale])
```

```python
scalar.transform(dataset[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scalar.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.head()
```

```python
for feature in data.columns:
    if data[feature].isnull().sum()>0:
        print("{} and {} NaN values".format(feature,data[feature].isnull().sum()))
```
