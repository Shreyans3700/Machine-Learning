```python
import pandas as pd
```

```python
import statsmodels.api as sm

df_adv = pd.read_csv("/Users/shreyanshpandey/Downloads/Multicollinearity-master/data/Advertising.csv")
```

```python
df_adv.head()
```

```python
df_adv.drop("Unnamed: 0",axis=1,inplace=True)
df_adv.head()
```

```python
X = df_adv.drop("sales",axis=1)
y = df_adv["sales"]
```

```python
X.head(),y.head()
```

```python
#fit a OLS model with an intercept on TV and radio
X = sm.add_constant(X)
model = sm.OLS(y,X).fit()
```

```python
model.summary()
```
