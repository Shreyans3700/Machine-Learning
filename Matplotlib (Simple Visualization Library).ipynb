```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
X = np.arange(0,10)
y = np.arange(11,21)

plt.scatter(X,y,c="r")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Graph")
```

```python
x = np.arange(0,11)
y = x * x
```

```python
plt.plot(x,y,c="r")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Parabolic Graph")

#to save the image in png
plt.savefig("Test.png")
```

```python
#creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(X,y,'b*')

plt.subplot(2,2,3)
plt.plot(x,y,'g')
```

```python
X = np.arange(0,11)
y = 2 * x * x + 5

plt.plot(X,y,c='b')
plt.xlabel("X label")
plt.ylabel("Y_label")
plt.title("Matplotlib demonstration")
```

```python
X = np.arange(0,4*np.pi,0.01)
y = np.sin(X)

plt.plot(X,y,c='b')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Sine Cirve")
```

```python
X = np.arange(0,4*np.pi,0.01)
y = np.sin(X)
z = np.cos(X)

plt.subplot(2,1,1)
plt.plot(X,y,'r')

plt.subplot(2,1,2)
plt.plot(X,z,'b')
```

```python
#bar plots
X = [2,3,4,5,6,7,8]
y = [7,3,1,7,8,4,2]

plt.bar(X,y)
```

```python
#histogram
arr = [22,34,12,56,78,22,3,4,34,56,78,12,56,22]
plt.hist(arr,bins=20)
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False); 
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
